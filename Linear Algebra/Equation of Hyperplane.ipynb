# Equation of Hyperplane

### Definition:
A hyperplane is a flat surface of dimension (n−1) in an n-dimensional space that can separate the space into two halves.
- In 2D, it’s a line
- In 3D, it’s a plane
- In nD, it’s a (n−1)-dimensional flat surface
  `(e.g., in 4D, it's a 3D space!)`

### Mathematical Definition of a Hyperplane
Let’s say we are in **n-dimensional space** and we have:
#### 𝑤⃗=[𝑤1,𝑤2,...,𝑤𝑛] `(weight vector)`

#### 𝑥⃗=[𝑥1,𝑥2,...,𝑥𝑛] `(input vector)`

#### 𝑏 = bias term `(like intercept)`

**Then the equation of a hyperplane is:**
#### 𝑤⃗⋅𝑥⃗+𝑏=0

**Or written out:**
#### 𝑤1𝑥1 + 𝑤2𝑥2 +⋯⋯+ 𝑤𝑛𝑥𝑛 + 𝑏=0

### Gemoetric Meaning
- This equation defines a boundary where the dot product of the input and weights, plus the bias, equals zero.
- Every point 𝑥⃗ that satisfies this equation lies on the hyperplane.
If the result is:
    - \>0: it's on one side of the hyperplane
    - <0: it's on the other side



### In Machine Learning: Why Hyperplanes Matter
Hyperplanes are `decision boundaries`.
They help us classify data points when we have more than two or three features (columns).

#### In a binary classification task:
The goal is to find a hyperplane that separates two classes.

`For example: "spam" vs "not spam", or "tumor" vs "no tumor"`.

#### Algorithms that use hyperplanes:
| Algorithm                           | Role of Hyperplane                                                        |
| ----------------------------------- | ------------------------------------------------------------------------- |
| **SVM (Support Vector Machine)**    | Finds the hyperplane with the **maximum margin** between two classes      |
| **Logistic Regression**             | Uses a linear hyperplane to **predict probabilities**, then thresholds it |
| **Perceptron / Linear Classifiers** | Uses it to determine class labels                                         |

### In Higher Dimensions (More Than 3 Columns)
In ML, data often has **10s or 100s of features**. You can’t visualize that directly.

But mathematically:
- A hyperplane still exists and it just divides the **n-dimensional space** into two halves.
- Algorithms like SVM **don’t need visualization**; they can compute the hyperplane **using vector math and optimization**.

Think of it like this:
- Even if you can’t draw a 100D space, the math still finds a hyperplane that splits it just like a 2D line does in a graph.

<br>

In simple terms, we can say that **The equation of a hyperplane is just the equation of a line or plane — generalized to higher dimensions**.

Let's compare:
| Dimension | Object     | Equation Form                              |
| --------- | ---------- | ------------------------------------------ |
| **2D**    | Line       | $w_1x_1 + w_2x_2 + b = 0$                  |
| **3D**    | Plane      | $w_1x_1 + w_2x_2 + w_3x_3 + b = 0$         |
| **nD**    | Hyperplane | $w_1x_1 + w_2x_2 + \dots + w_nx_n + b = 0$ |

### Example for clarity:
### In 2D:
#### `3𝑥+2𝑦−6=0`
**-> This is a line.**

### In 3D:
#### `3𝑥+2𝑦+𝑧−6=0`
**-> This is a plane.**

### In 100D:
#### `𝑤1𝑥1+𝑤2𝑥2+⋯+𝑤100𝑥100+𝑏=0`
**-> This is a hyperplane.**

## Let's understand `The Equation of Hyperplane` with `Linear Regression`.

In linear regression with more than 3D (i.e., more than 2 features),` the model's prediction is represented by a hyperplane in n-dimensional space.

### Linear Regression Formula Recap:
Let:

- #### 𝑥1,𝑥2,...,𝑥𝑛 `be features`

- #### 𝑤1,𝑤2,...,𝑤𝑛 `be weights (coefficients)`

- #### 𝑏 `be bias (intercept)`

- #### 𝑦 `be the predicted value`

#### Then the model is:

#### 𝑦 = 𝑤1𝑥1+𝑤2𝑥2+⋯+𝑤𝑛𝑥𝑛+𝑏

`This is exactly **the equation of a hyperplane** in **(n+1)-dimensional space** — the **𝑛** input dimensions and the output.`

<br>

| No. of Features | Regression Model Shape | Equation Type                     |
| --------------- | ---------------------- | --------------------------------- |
| 1 feature       | Line in 2D             | $y = w_1x_1 + b$                  |
| 2 features      | Plane in 3D            | $y = w_1x_1 + w_2x_2 + b$         |
| 3+ features     | Hyperplane in 4D+      | $y = w_1x_1 + \dots + w_nx_n + b$ |


### So, in linear regression:
- The model equation defines a hyperplane

- We fit this hyperplane so it minimizes the error between predicted values 𝑦^ and true values 𝑦

- We can't visualize it beyond 3D, but the math works perfectly in any dimension

<br>

## Now, let's understand it by using a classic dataset called **MNIST Dataset**
The `MNIST dataset` is a classic example of working in `high-dimensional vector space`, and it's a perfect context to understand how hyperplanes and vector operations scale to many dimensions.

### What is the MNIST Dataset?
It contains images of handwritten digits (0 to 9)

Each image is 28 × 28 pixels
→ So each image is flattened into a vector of 784 features

#### 28 × 28 = 784

So, each image is a point in 784-dimensional space:

#### 𝑥⃗=[𝑥1,𝑥2,…,𝑥784]

<br>

#### How Do We Formulate It?
Suppose we're using Linear Models (e.g., Logistic Regression or SVM) to classify digits:

Let’s say:

- 𝑥⃗
  is the input vector (an MNIST image) with 784 pixel values

- 𝑤⃗
  is the weight vector (one per class in multiclass classification)

- 𝑏
 is the bias (scalar)

- 𝑦^
  is the output (logit score or class probability)


Then the model equation becomes:

#### 𝑦^=𝑤⃗⋅𝑥⃗+𝑏=𝑤1𝑥1+𝑤2𝑥2+⋯+𝑤784𝑥784+𝑏

--- 

### What's Happening Geometrically?
- Each digit image is a point in 784-dimensional space

- Each class (0–9) has its own set of weights and bias → its own hyperplane

- The model computes:
    - Which side of each hyperplane the input falls on
    - Or: Which class hyperplane has the highest score

This is how it separates the digits even though we can't visualize 784D.

---
#### For Multiclass (0–9):
Instead of a single hyperplane, we use one per class.

That means:

- #### 𝑤⃗ (0),𝑤⃗ (1),…,𝑤⃗ (9): 10 weight vectors

- Model computes 10 scores:

#### 𝑧𝑖=𝑤⃗ (𝑖)⋅𝑥⃗+𝑏(𝑖) for 𝑖=0 to 9

- Then picks the class with the highest score (or highest softmax probability)

<br>

### Summary:
#### MNIST and Hyperplanes
- MNIST images: 784-dimensional vectors (28×28 pixels)
- A linear model fits:

  **𝑦^=𝑤⃗⋅𝑥⃗+𝑏**

  **where, 𝑥⃗∈𝑅784**

- Each class (digit) has its own weight vector and hyperplane
- The model classifies by checking which side of each hyperplane the input lies
- In multiclass, it calculates 10 outputs (one per digit) and picks the highest


### Hyperplanes in High Dimensions
- Hyperplanes still work in 784D just like lines in 2D or planes in 3D
- ML models like Logistic Regression and SVM learn those hyperplanes
- These hyperplanes separate digits in high-dimensional space



<br>

### 1. Visualize a Line in 2D

```python
import matplotlib.pyplot as plt
import numpy as np

# Equation: w1*x + w2*y + b = 0 → solve for y
w1, w2, b = 2, -3, -6

x = np.linspace(-10, 10, 100)
y = (-w1 * x - b) / w2  # solving for y

plt.plot(x, y, label='Hyperplane (Line)')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("2D Line (Hyperplane in 2D)")
plt.grid(True)
plt.legend()
plt.show()
```

### 2. Visualize a Plane in 3D

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Equation: w1*x + w2*y + w3*z + b = 0 → solve for z
w1, w2, w3, b = 1, 2, -1, -5

x = np.linspace(-10, 10, 10)
y = np.linspace(-10, 10, 10)
X, Y = np.meshgrid(x, y)
Z = (-w1 * X - w2 * Y - b) / w3  # solving for z

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7, color='cyan', edgecolor='gray')
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("x3")
ax.set_title("3D Plane (Hyperplane in 3D)")
plt.show()
```

### 3. Conceptual Hyperplane in Higher Dimensions (nD)

Since we can’t visualize 784D, we simulate the idea by:

- Creating random nD data
- Projecting it into 2D using PCA
- Showing a decision boundary (projected hyperplane)

```python
from sklearn.datasets import make_classification
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Generate synthetic high-dimensional data (e.g., 10D)
X, y = make_classification(n_samples=300, n_features=10, n_informative=2, n_redundant=0, n_classes=2)

# Fit logistic regression (linear model → hyperplane)
model = LogisticRegression()
model.fit(X, y)

# Reduce dimensions to 2D using PCA for visualization
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

# Decision boundary (project hyperplane)
w = model.coef_[0]
b = model.intercept_[0]

# Project weight vector to 2D space
w_2d = pca.transform([w])[0]

# Plot
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y, cmap='coolwarm', alpha=0.6)
x_vals = np.linspace(X_2d[:,0].min(), X_2d[:,0].max(), 100)
y_vals = -(w_2d[0] * x_vals + b) / w_2d[1]  # derived line
plt.plot(x_vals, y_vals, color='black', label='Projected Hyperplane')
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Hyperplane in High-Dimensional Space (PCA Projection)")
plt.legend()
plt.grid(True)
plt.show()
```
